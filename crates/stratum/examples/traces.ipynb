# Traces — the partial-order view

A protocol has many *runs* but often far fewer *traces*. `#traces` enumerates the traces — partial orders of events — collapsing the incidental order of independent steps while keeping genuine branches. Bind the set with `-> tr` and address one element as `tr[i]`.

```stratum
#traces new a, b
a!(0) | a(x).0 | b!(0) | b(y).0 -> tr
```

**One** trace: the two reactions touch disjoint channels, so their order is incidental. The `∥` reads *concurrent*. Look at the partial order:

```stratum
#trace tr[0]
```

Its two interchangeable linearizations — the runs the single trace stands for:

```stratum
#lin tr[0]
```

What an agent watching only channel `a` can see — the projection keeps just its events, under the induced order:

```stratum
#project tr[0] a
```

A **race** — one output, two receivers competing — branches into *distinct* traces instead:

```stratum
#traces new c
c!(0) | c(x).0 | c(y).0 -> rr
```
